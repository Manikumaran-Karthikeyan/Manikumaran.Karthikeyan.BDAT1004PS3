Question 6

Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

US_Marriages_And_Divorces = pd.read_csv("us-marriages-divorces-1867-2014.csv")
US_Marriages_And_Divorces.plot.line(x= "Year", y=["Marriages_per_1000", "Divorces_per_1000"]) 
plt.xlabel('Years') 
plt.ylabel('Marriages and Divorces') 
plt.title("Number of Marriages and Divorces per capita in USA between 1867-2014") 
plt.legend()
```

Question 7

Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

#filtering data based on year and 
US_Marriages_And_Divorces = US_Marriages_And_Divorces[(US_Marriages_And_Divorces.Year == 1900) | (US_Marriages_And_Divorces.Year == 1950) | (US_Marriages_And_Divorces.Year == 2000)]
US_Marriages_And_Divorces = US_Marriages_And_Divorces.drop(columns = ['Marriages', 'Divorces', 'Population'])
US_Marriages_And_Divorces = US_Marriages_And_Divorces.set_index('Year')
US_Marriages_And_Divorces
```

```python
US_Marriages_And_Divorces.plot.bar() 
plt.xlabel('Years: 1900,1950 and 2000') 
plt.ylabel('Number of Marriages & Divorces') 
plt.title("Number of marriages & Divorces per capita in USA between 1900, 1950, and 2000") 
plt.legend() 
```

Question 8

Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
the actors by their kill count and label each bar with the corresponding actor's name.
Don't forget to label your axes!


```python
import pandas as pd
import matplotlib.pyplot as plt

Actor_kill_counts = pd.read_csv("actor_kill_counts.csv")
colours = ['#F8B195','#F67280','#C06C84','#6C5B7B','#355C7D']
Actor_kill_counts['Count'].sort_values().plot.barh(color=colours)

plt.xlabel('KillCounts') 
plt.ylabel('Actors') 
plt.title("Actor Kill Counts") 
plt.legend()
```
